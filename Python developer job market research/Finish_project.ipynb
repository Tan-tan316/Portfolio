Завдання 7. Географічний аналіз компенсації Python-розробників

```python
df_country_comp=df_py.groupby('Country')[['ConvertedCompYearly']].mean().round(2)
```

```python
df_country_comp=df_country_comp.rename(columns={'ConvertedCompYearly':'MeanCompYearly'})
```

```python
df_country_comp['MedianCompYearly']=df_py.groupby('Country')[['ConvertedCompYearly']].median()
```

```python
df_country_comp=df_country_comp.reset_index().dropna()
```

```python
df_country_comp
```

```python
country_counts = df_py['Country'].value_counts()

valid_countries = country_counts[country_counts >= 50].index
df_filtered = df_py[df_py['Country'].isin(valid_countries)]

top10_filtered = df_filtered.groupby('Country')['ConvertedCompYearly'].mean().sort_values(ascending=False).head(10)
```

```python
top10_filtered
```

```python
country_counts = df['Country'].value_counts()

valid_countries = country_counts[country_counts >= 50].index
df_filtered = df_country_comp[df_country_comp['Country'].isin(valid_countries)]

top10 = df_filtered.sort_values(by='MeanCompYearly', ascending=False).head(10)
top10['Country'] = top10['Country'].replace({
    'United Kingdom of Great Britain and Northern Ireland': 'UK',
    'United States of America': 'USA'})

plt.figure(figsize=(10,6))
bars = plt.bar(top10['Country'], top10['MeanCompYearly'], color='dodgerblue', label='Mean')
plt.yscale('log')

for i, country in enumerate(top10['Country']):
    median = top10.loc[top10['Country']==country, 'MedianCompYearly'].values[0]
    plt.hlines(median, i-0.4, i+0.4, colors='orange', label='Median' if i==0 else "", linewidth=2)

plt.gca().yaxis.set_major_formatter(NullFormatter())
plt.gca().yaxis.set_minor_formatter(NullFormatter())

plt.xticks(rotation=45, ha='right')
plt.ylabel('Compensation')
plt.title('Топ-10 країн (>=5 респондентів): Mean (стовпці) і Median (лінія)')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}', ha='center', va='bottom', fontsize=9)

plt.legend()
plt.tight_layout()
plt.show()
```

*Завдання 9. Аналіз популярності Python по віковим категоріям 

```python
age_all=df['Age'].value_counts().reset_index().rename(columns={'count':'count_all'})
age_all.set_index('Age', inplace=True)
age_all
```

```python
age_all['count_python']=df_py['Age'].value_counts()
age_all=age_all.reset_index()
```

```python
age_all
```

```python
age_all['percent_python']=round(age_all['count_python']/age_all['count_all']*100,2)
```

```python
age_all = age_all.sort_values('Age')
```

```python
age_all
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.barh(age_all['Age'], age_all['percent_python'], color='skyblue')
plt.xlabel('Percent using Python (%)')
plt.title('Популярність Python по віковим категоріям')

for i, val in enumerate(age_all['percent_python']):
    plt.text(val + 0.5, i, f'{val:.1f}%', va='center')

plt.xlim(0, 50)  
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.savefig('age_python', dpi=300, bbox_inches='tight')
plt.show()
```

*Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
df['ConvertedCompYearly'].quantile(0.75)
```

```python
mask=df['ConvertedCompYearly'] > df['ConvertedCompYearly'].quantile(0.75)
```

```python
df_new=df[mask]
```

```python
mask2=df_new['RemoteWork'] == 'Remote'
```

```python
df_new=df_new[mask2]
```

```python
df_ind=df_new['Industry'].value_counts().reset_index()
```

```python
df_ind
```

```python

df_plot = df_ind.sort_values("count", ascending=True)

plt.figure(figsize=(14, 10))
bars = plt.barh(df_plot["Industry"], df_plot["count"])

plt.xlabel("Кількість респондентів")
plt.ylabel("Індустрія")
plt.title("Розподіл респондентів за індустріями")

# додаємо підписи
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 5,                    
        bar.get_y() + bar.get_height()/2,
        f"{int(width)}",
        va="center"
    )

plt.tight_layout()
plt.savefig('industrias', dpi=300, bbox_inches='tight')
plt.show()
```

Якщо рахувати що найрозповсюдженіші індустрії серед високооплачуваних віддалених працівників це ті кількість респондентів в яких більша за 100, то це: "Software Development" , "Fintech" , "Healthcare" , "Internet, Telecomm or Information Services".
